## Parte B

Un *Vector Network Analyzer* (VNA) es un instrumento de medición utilizado para caracterizar el comportamiento de dispositivos electrónicos en función de la frecuencia. Su función principal es analizar cómo una señal eléctrica se transmite y se refleja al interactuar con un *dispositivo bajo prueba* (DUT), permitiendo obtener tanto la magnitud como la fase de dichas señales.Ademas, el VNA mide los parámetros S.

El uso de un VNA resulta fundamental en situaciones donde se trabaja con altas frecuencias (RF y microondas), ya que en estos casos los efectos de reflexión y desadaptación de impedancias son críticos.

La aplicacion de un VNA es fundamental en situaciones como: **Diseño y ajuste de antenas, caracterización de filtros, análisis de líneas de transmisión, desarrollo de circuitos de RF.**
En estos casos, el VNA resulta indispensable porque permite observar con precisión cómo interactúan las señales dentro del sistema, algo que no puede lograrse únicamente con mediciones de tensión o potencia por ejmeplo.

------------------------------------------------------------------------------------------------------------------------------------------

A partir de la información del *NanoVNA* y del osciloscopio digital *GW Instek GDS-1000A-U*, se pueden comparar tres características clave:

1. Ancho de banda:
   NanoVNA: Opera en un rango de frecuencias mucho más alto, típicamente hasta varios GHz (dependiendo la versión, hasta ~3–4.4 GHz).
   GW Instek: Presenta anchos de banda de 70 MHz, 100 MHz y 150 MHz

   El NanoVNA tiene un ancho de banda significativamente mayor, ya que está diseñado para aplicaciones de RF y microondas, mientras que el DSO está orientado a señales de menor frecuencia
   
2. Rango dinamico:
   NanoVNA: Posee un rango dinámico alto (de ~90 dB en versiones modernas), lo que permite medir con precisión diferencias entre señales reflejadas y transmitidas.
    GW Instek: No está optimizado para rango dinámico en términos de potencia RF. Su resolución y precisión están limitadas por resolución vertical o ruido del sistema.

    El NanoVNA tiene un mayor rango dinámico efectivo en frecuencia, fundamental para mediciones de parámetros S, mientras que el GW no está diseñado para este tipo de análisis.


3. Impedancia de entrada
    NanoVNA: Tiene impedancia característica de 50 Ω, ya que trabaja con sistemas de RF donde la adaptación de impedancia es crítica.
    GW Instek: Presenta una impedancia de 1 MΩ ±2%

    El GW tiene alta impedancia de entrada para no cargar el circuito, mientras que el NanoVNA usa 50 Ω para garantizar adaptación en líneas de transmisión.


Si bien tanto el NanoVNA como el DSO GW Instek GDS-1000A-U son instrumentos de medición, no son directamente comparables en términos funcionales, ya que son utilizados para análisis completamente distintos: dominio de la frecuencia vs. dominio del tiempo.

------------------------------------------------------------------------------------------------------------------------------------------

Además de las mediciones de parámetros S en el dominio de la frecuencia, el NanoVNA V2 también cuenta con capacidad de análisis en el dominio del tiempo.

Según el manual del equipo, esta funcionalidad se implementa mediante una transformación matemática (similar a una Transformada Inversa de Fourier), que permite convertir los datos medidos en frecuencia a una representación en el tiempo.

Esta herramienta se utiliza principalmente en modo TDR (Time Domain Reflectometry), lo que permite:

- Identificar discontinuidades en líneas de transmisión
- Estimar la distancia a fallas o desadaptaciones
-  Analizar conectores, cables y empalmes

En este modo, el NanoVNA no mide directamente en el dominio del tiempo (como lo haría un osciloscopio), sino que reconstruye la respuesta temporal a partir de las mediciones en frecuencia.
